```python
# Exercise 4.1: Analyzing grade data
# TODO: Complete this analysis

# Sample student grades
grades = [85, 92, 78, 90, 88, 95, 82, 91, 87, 94]

# TODO: Calculate the average grade
average = sum(grades) / len(grades)

# TODO: Find the highest grade
highest = max(grades)

# TODO: Find the lowest grade  
lowest = min(grades)

# TODO: Count how many students got A (90+)
num_a_grades = sum(1 for grade in grades if grade >= 90)

# Print results
print(f"Class Statistics:")
print(f"  Average: {average:.2f}")
print(f"  Highest: {highest}")
print(f"  Lowest: {lowest}")
print(f"  Number of A grades: {num_a_grades}")
```

```python
# Exercise 4.3: Create a useful function
# TODO: Write a function that converts letter grade to GPA

def letter_to_gpa(letter_grade):
    """
    Convert a letter grade to GPA points.
    
    Args:
        letter_grade: A string like 'A', 'B', 'C', 'D', 'F'
    
    Returns:
        GPA points as a float (A=4.0, B=3.0, C=2.0, D=1.0, F=0.0)
    """
    # TODO: Implement this function
    if letter_grade == 'A':
        return 4.0
    elif letter_grade == 'B':
        return 3.0
    elif letter_grade == 'C':
        return 2.0
    elif letter_grade == 'D':
        return 1.0
    elif letter_grade == 'F':
        return 0.0
    else:
        raise ValueError("Invalid letter grade")


# Test your function
print(f"A = {letter_to_gpa('A')} GPA")
print(f"B = {letter_to_gpa('B')} GPA")
print(f"C = {letter_to_gpa('C')} GPA")
```
